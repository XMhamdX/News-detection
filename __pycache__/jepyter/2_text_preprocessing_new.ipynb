# معالجة النصوص
هذا الملف يحتوي على كود معالجة وتنظيف النصوص من البيانات المجمعة

```python
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# تحميل البيانات المطلوبة من NLTK
nltk.download('punkt')
nltk.download('stopwords')
```

```python
def clean_text(text):
    # إزالة علامات HTML
    text = re.sub('<[^>]*>', '', str(text))
    
    # تحويل النص إلى أحرف صغيرة
    text = text.lower()
    
    # إزالة الأرقام والرموز الخاصة
    text = re.sub('[^a-zA-Z\s]', '', text)
    
    # تقسيم النص إلى كلمات
    words = word_tokenize(text)
    
    # إزالة الكلمات الشائعة
    stop_words = set(stopwords.words('english'))
    words = [word for word in words if word not in stop_words]
    
    # إعادة تجميع النص
    text = ' '.join(words)
    
    # إزالة المسافات الزائدة
    text = re.sub('\s+', ' ', text).strip()
    
    return text
```

```python
# قراءة البيانات
data = pd.read_csv('bbc-text.csv')

# التحقق من أسماء الأعمدة
print("أسماء الأعمدة في الملف:")
print(data.columns)

# عرض البيانات قبل المعالجة
print("\nعينة من البيانات قبل المعالجة:")
print(data.head())
```

```python
# تطبيق التنظيف على النصوص
data['Content'] = data['Content'].apply(clean_text)

# عرض النتائج بعد المعالجة
print("\nعينة من النصوص بعد المعالجة:")
print(data['Content'].head())

# حفظ البيانات المعالجة
data.to_csv('cleaned_bbc_text.csv', index=False)
print("\nتم حفظ البيانات المعالجة في ملف 'cleaned_bbc_text.csv'")
```
